```python
import datetime

dt = datetime.datetime(2022, 6, 1, 7,0,0)
end = datetime.datetime(2022, 7, 31, 23, 59, 59)
step = datetime.timedelta(minutes=2)

result = []

while dt < end:
    result.append(dt.strftime('%Y-%m-%d %H:%M:%S'))
    dt += step
result
df = pd.DataFrame(result)
df.columns =['date']
df.to_csv('descriptif.csv',sep=';',index=False)
#2022-07-26 22:45:00
```
