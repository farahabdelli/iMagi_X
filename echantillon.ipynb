```python
import pandas
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

# data Johnny Alves

```python
#importer les données

data = pd.read_csv('data/stats Johnny alves.csv',sep=";")
data
```

```python
data = pd.read_csv('data/stats amaury sion.csv',sep=";")
data
```

```python
#choix des colonnes
data2 = data[["Date",	"airTemp",	"heaterSetpoint",	"orp",	"ph","thirdTemp",	"waterTemp"]]

#intervalle de date 
data2['Date'] = pd.to_datetime(data2['Date'])  
start_date = '14/07/2022 07:00'
end_date   = '14/09/2022 20:00'
mask = (data2['Date'] > start_date) & (data2['Date'] <= end_date)
data2 = data2.loc[mask]

#ajouter une colonne type pour indique le type de chaque valeur
data2air = data2[["Date",	"airTemp"]]
data2air = data2air.rename(columns={'Date': 'Date', 'airTemp': 'value'})
data2air=data2air.set_index("Date")
data2air['type']='airTemp'

data2heater = data2[["Date","heaterSetpoint"]]
data2heater = data2heater.rename(columns={'Date': 'Date', 'heaterSetpoint': 'value'})
data2heater=data2heater.set_index("Date")
data2heater['type'] = 'heaterSetpoint'

data2orp = data2[["Date","orp"]]
data2orp = data2orp.rename(columns={'Date': 'Date', 'orp': 'value'})
data2orp=data2orp.set_index("Date")
data2orp['type'] = 'orp'

data2ph = data2[["Date",	"ph"]]
data2ph = data2ph.rename(columns={'Date': 'Date', 'ph': 'value'})
data2ph=data2ph.set_index("Date")
data2ph['type'] = 'ph'

data2third = data2[["Date","thirdTemp"]]
data2third = data2third.rename(columns={'Date': 'Date', 'thirdTemp': 'value'})
data2third=data2third.set_index("Date")
data2third['type'] = 'thirdTemp'

data2water = data2[["Date","waterTemp"]]
data2water = data2water.rename(columns={'Date': 'Date', 'waterTemp': 'value'})
data2water=data2water.set_index("Date")
data2water['type'] = "waterTemp"

df_type = pandas.concat([data2air,data2heater,data2heater,data2orp,data2ph  ,data2third ,data2water ,],axis=0)
df_type = df_type.reset_index()

#datetime entre 8h et 18h
df_type['Date'] = pd.to_datetime(df_type['Date'])  
df=df_type[(df_type['Date'].apply(lambda x : x.hour)>=8) & (df_type['Date'].apply(lambda x : x.hour)<=18)]


#moyenne des valeurs chaque 2h
df['time'] = df['Date'].dt.round('120min')
aggregation = {'valeur': ('value', 'mean')}
df_agg = df.groupby(['time','type']).agg(**aggregation).reset_index()


#afficher le graphique
import plotly.express as px

fig = px.line(df_agg, x="time", y="valeur",color='type')
             


fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
         
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)

fig.show()
```
