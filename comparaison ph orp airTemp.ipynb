# data Johnny Alves

```python
#importer les données

datajohnny = pd.read_csv('data/johny2.csv',sep=";")
datajohnny
```

```python
#choix des colonnes
data2johnny = datajohnny[["Date",	"airTemp",	"heaterSetpoint",	"orp",	"ph",	"waterTemp"]]

#intervalle de date 
data2johnny['Date'] = pd.to_datetime(data2johnny['Date'])  
start_date = '14/07/2022 07:00'
end_date   = '14/09/2022 20:00'
mask = (data2johnny['Date'] > start_date) & (data2johnny['Date'] <= end_date)
data2 = data2johnny.loc[mask]

#ajouter une colonne type pour indique le type de chaque valeur
data2airjohnny = data2johnny[["Date",	"airTemp"]]
data2airjohnny = data2airjohnny.rename(columns={'Date': 'Date', 'airTemp': 'value'})
data2airjohnny=data2airjohnny.set_index("Date")
data2airjohnny['type']='airTemp'

data2heater = data2johnny[["Date","heaterSetpoint"]]
data2heater = data2heater.rename(columns={'Date': 'Date', 'heaterSetpoint': 'value'})
data2heater=data2heater.set_index("Date")
data2heater['type'] = 'heaterSetpoint'

data2orpjohnny = data2johnny[["Date","orp"]]
data2orpjohnny = data2orpjohnny.rename(columns={'Date': 'Date', 'orp': 'value'})
data2orpjohnny=data2orpjohnny.set_index("Date")
data2orpjohnny['type'] = 'orp'

data2phjohnny = data2johnny[["Date",	"ph"]]
data2phjohnny = data2phjohnny.rename(columns={'Date': 'Date', 'ph': 'value'})
data2phjohnny=data2phjohnny.set_index("Date")
data2phjohnny['type'] = 'ph'

data2water = data2johnny[["Date","waterTemp"]]
data2water = data2water.rename(columns={'Date': 'Date', 'waterTemp': 'value'})
data2water=data2water.set_index("Date")
data2water['type'] = "waterTemp"

df_typejohnny = pandas.concat([data2airjohnny,data2heater,data2heater,data2orpjohnny,data2phjohnny  ,data2water ,],axis=0)
df_typejohnny = df_typejohnny.reset_index()

#datetime entre 8h et 18h
df_typejohnny['Date'] = pd.to_datetime(df_typejohnny['Date'])  
dfjohnny=df_typejohnny[(df_typejohnny['Date'].apply(lambda x : x.hour)>=8) & (df_typejohnny['Date'].apply(lambda x : x.hour)<=18)]


#moyenne des valeurs chaque 2h
dfjohnny['time'] = dfjohnny['Date'].dt.round('120min')
aggregation = {'valeur': ('value', 'mean')}
df_aggjohnny = dfjohnny.groupby(['time','type']).agg(**aggregation).reset_index()


#afficher le graphique
import plotly.express as px

fig = px.line(df_aggjohnny, x="time", y="valeur",color='type')
             


fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
         
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)

fig.show()
```
